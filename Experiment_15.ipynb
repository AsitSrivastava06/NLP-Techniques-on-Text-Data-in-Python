```python
#steaming
#steaming reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["eating", "playing", "fighting","watching"]
for w in words:
  print(stemmer.stem(w))
```

```python
# Lemmatization Theory

'Lemmatization is an advanced text normalization technique in Natural Language Processing (NLP) that aims to reduce words to their base or dictionary form, known as a **lemma**.'

### How it Works
'''Unlike **stemming** (which often simply chops off word endings and can result in non-words), lemmatization is:

*   **Context-Aware:** It analyzes the word's part of speech (e.g., noun, verb) and its surrounding context to derive the correct lemma.
*   **Dictionary-Based:** It uses a vocabulary (like WordNet in NLTK) and morphological analysis to ensure the output is always a valid word from the dictionary.'''

### Key Characteristics
'''Accuracy:** Generally more accurate than stemming because it considers linguistic context.
Valid Output:** Always returns a grammatically correct base form.
Computational Cost:** More computationally intensive than stemming due to the need for dictionary lookups and linguistic rules.'''

### Purpose and Importance
'''The main goal of lemmatization is to normalize text data by grouping different inflected forms of a word into a single base form. For example:
*   'runs', 'running', 'ran' all become **'run'**
*   'better', 'best' both become **'good'**'''
'''
This normalization is crucial for various text analysis tasks, including:
*   **Information Retrieval:** Helps search engines find relevant documents regardless of word inflection.
*   **Text Classification:** Reduces feature space and improves model accuracy by treating different forms of a word as the same underlying concept.
*   **Machine Translation:** Ensures consistency in word representation across languages.'''
```

```python
#Lemmatization
#Lemmatization converts words to their dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("fighting"))
print(lemmatizer.lemmatize("playing"))
```
